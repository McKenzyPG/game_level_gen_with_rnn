```
!pip install -q textgenrnn
from google.colab import files
from textgenrnn import textgenrnn
import os
```

```
uploaded = files.upload()
all_files = [(name, os.path.getmtime(name)) for name in os.listdir()]
latest_file = sorted(all_files, key=lambda x: -x[1])[0][0]
```

```
textgen = textgenrnn()
```

```
textgen.reset()
```

```
textgen.train_from_file(latest_file,new_model=True,
                        rnn_bidirectional=True,
                        rnn_size=128,
                        dim_embeddings=100,
                        num_epochs=1000)
```

```
print(textgen.model.summary())
```

```
textgen.generate(24, temperature = 1.0)
```

```
textgen.generate_to_file('textgenrnn_maps.txt', n=5)
```

```
textgen.save('model.hdf5')
```
